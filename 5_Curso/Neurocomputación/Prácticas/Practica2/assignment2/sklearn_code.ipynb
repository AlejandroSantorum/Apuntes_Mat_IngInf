# XOR

```python
sets = read2('./test_files/xor.txt')

xtrain, ytrain, xtest, ytest = sets
```

```python
xtrain = np.asarray(xtrain)
print(xtrain.shape)
ytrain = np.ravel(np.array(ytrain))
print(ytrain.shape)
xtest = np.asarray(xtest)
print(xtest.shape)
ytest = np.ravel(np.array(ytest))
print(ytest.shape)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(8), activation='logistic', alpha=0, solver='sgd', tol=0,
                    learning_rate='constant', learning_rate_init=0.1, max_iter=1000, batch_size=1,
                    nesterovs_momentum=False, momentum=0).fit(xtrain, ytrain)
```

```python
clf.score(xtest, ytest)
```

```python
clf.predict(xtest)
```

```python
plt.plot(clf.loss_curve_)
```

```python
# https://stackoverflow.com/questions/44896453/training-a-neural-network-to-compute-xor-in-scikit-learn
# Changing activation func and solver

clf = MLPClassifier(hidden_layer_sizes=(8), activation='relu', alpha=0, solver='adam', tol=0.01,
                    learning_rate='constant', learning_rate_init=0.1, max_iter=1000, batch_size=1,
                    nesterovs_momentum=False, momentum=0).fit(xtrain, ytrain)

clf.score(xtest, ytest)
```

```python
clf.predict(xtest)
```

```python
plt.plot(clf.loss_curve_)
```

<h3>Fibonacci recursivo a iterativo</h3>

```python
def fibon3(n):
    L = [1,1]
    while len(L) <= n+1:
        L.append(L[-2]+L[-1])
    return L
```

```python
fibon3(10)
```

```python
time RES = fibon3(10^5)
```

<h3>Suma de los cubos:</h3>

```python
def suma(n):
    suma = 0
    j = 1
    while j < n+1:
        suma += j^3
        j += 1
    return suma
```

```python
suma(100)
```

```python
def suma2(n):
    if n == 1:
        return 1
    else:
        return suma2(n-1)+n^3
```

```python
suma2(100)
```

```python
time suma(10^4)
```

```python
sys.setrecursionlimit(10^4+100)
%time suma2(10^4)
```

<h3>Factorial:</h3>

```python
def fact(n):
    prod = 1
    for i in xsrange(2,n+1):
        prod *= i
    return prod
```

```python
fact(5)
```

<h3>M&aacute;ximo Com&uacute;n Divisor:</h3>

```python
def mcd(n,m):
    while m > 0:
        n,m = m,n%m
    return n
```

```python
mcd(5*87665,77774*5)
```

```python
def mcd2(n,m):
    '''Maximo com\'un divisor de dos enteros positivos n,m con n>m'''
    if m == 0:
        return n
    else:
        return mcd(m,n%m)
```

```python
mcd2(5*87665,77774*5)
```

<h4>Recursi&oacute;n e inducci&oacute;n</h4>
<p>Una soluci&oacute;n al primer ejercicio propuesto dentro del ejercicio 9 de la lista al final del cap&iacute;tulo 4 de las notas.</p>

```python
def comprueba(N):
    var('n')
    L = []
    S = 1
    F(n) = n*(n+1)*(2*n+1)/6
    if N == 1:
        if F(n=1) == 1:
            L.append(1)
            return S,L
        else:
            L.append(0)
            return S,L

    else:
        S,L= comprueba(N-1)
        S += N^2
        if S == F(n=N):
            L.append(1)
        else:
            L.append(0)
        return S,L
```

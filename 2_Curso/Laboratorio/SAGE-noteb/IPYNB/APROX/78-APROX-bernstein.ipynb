```python
def B(n,f):
    var('t')
    return sum([binomial(n,p)*f(x=(p/n).n())*(1-t)^(n-p)*t^p for p in srange(n+1)])
```

```python
f(x)=exp(x);
expand(B(15,f))
```

```python
plot(f,0,1)+plot(B(100,f),0,1)
```

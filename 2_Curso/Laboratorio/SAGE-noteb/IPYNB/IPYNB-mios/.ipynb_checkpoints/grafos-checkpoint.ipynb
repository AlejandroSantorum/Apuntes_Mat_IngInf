```python
def cuadrados_mod(n,m):
    return (n*n)%m

def diccionario(m):
    dicc = {}
    for int1 in srange(m):
        int22 = cuadrados_mod(int1, m)
        dicc[int1] = [int22]
    return dicc

d = diccionario(7)
d
```

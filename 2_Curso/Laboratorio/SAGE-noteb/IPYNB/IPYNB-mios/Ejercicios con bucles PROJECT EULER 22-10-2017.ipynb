EJERCICIO 1 - Existen enteros, por ejemplo 145, que son iguales a la suma de los factoriales de sus dígitos. Determina todos los enteros con esta propiedad.

```python
for n in srange(1, 1000000):
    L = n.digits()
    l = len(L)
    a=0
    for i in xsrange (0,l):
        a += factorial(L[i])

    if n==a:
        print(n)
```

EJERCICIO 4(Project Euler) - A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
LR = list()

for i in xsrange(100, 1000):
    for j in xsrange(100, 1000):
        a = i*j  #calculamos número a evaluar si es un palíndromo o no
        
        L = a.digits()
        LL=list(L)
        LL.reverse()
        flag = 0
        l=len(L)
        for k in xsrange(0, l):
            if L[k] != LL[k]:
                flag = 1
                break
            
        if flag == 0:
            LR.append(a)

ll = len(LR)
maximo = 0
for m in xsrange(0, ll):
    if maximo < LR[m]:
        maximo = LR[m]
        
print("Palíndromo más grande que cumpla los requisitos pedidos:"+str(maximo))
print(" ")
```

EJERCICIO 421(Proyect Euler) - Numbers of the form n15+1 are composite for every integer n > 1.
For positive integers n and m let s(n,m) be defined as the sum of the distinct prime factors of n15+1 not exceeding m.

E.g. 215+1 = 3×3×11×331.
So s(2,10) = 3 and s(2,1000) = 3+11+331 = 345.

Also 1015+1 = 7×11×13×211×241×2161×9091.
So s(10,100) = 31 and s(10,1000) = 483.
Find ∑ s(n,108) for 1 ≤ n ≤ 1011.

```python
def s(n, tope):                  #Función s(n,m) definida en ProjectEuler
    nn = n^15+1
    L = list(nn.factor())
    l = len(L)
    LL = list()
    for i in xsrange(0, l):
        a = L[i][0]
        if a <= tope:
            LL.append(a)
        elif a > tope:
            break
    
    ll=len(LL)
    suma=0
    for j in xsrange(0, ll):
        suma += LL[j]
    return suma


#Lo que se pide calcular en ProjectEuler es demasiado grande. La solución de abajo es con número más pequeños

total = 0
for n in xsrange(1, 1001):
    total += s(n, 100)
    
print("Resultado:"+str(total))      
```

```python
def actualiza_urna(L):
    b,n=L[0],L[1]
    x = randint(1,b+n)
    if x<=b:
        return [b+1,n]
    else:
        return [b,n+1]
```

```python
actualiza_urna(actualiza_urna([1,3]))
```

```python
def camino(L,n):
    LL = [L]
    for int in srange(n):
        LL.append(actualiza_urna(LL[-1:][0]))
    return LL
```

```python
camino([1,1],10)
```

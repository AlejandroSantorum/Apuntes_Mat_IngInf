<h3>Cuerdas-Ej-11</h3>
<p>Usando un poco de trigonometr&iacute;a se puede ver que la longitud del lado del tri&aacute;ngulo equil&aacute;tero inscrito en la circunferencia de radio $1$ es $\sqrt{3}$, y tambi&eacute;n que las cuerdas que miden m&aacute;s de $\sqrt{3}$ son exactamente las que su punto medio dista del origen menos que $1/2$.</p>

```python
def punto():
    t = 2*pi.n()*random()
    P = cos(t),sin(t)
    return P
```

```python
punto()
```

```python
def punto_medio():
    P,Q = punto(),punto()
    return (P[0]+Q[0])/2,(P[1]+Q[1])/2
```

```python
punto_medio()
```

```python
def prob_cuerda1(R):
    cont = 0
    for muda in xsrange(R):
        P = punto_medio()
        if P[0]^2+P[1]^2 < 1/4:
            cont += 1
    return (cont/R).n()
```

```python
prob_cuerda1(10^5)
```

```python
def grafica1(R):
    points([punto_medio() for muda in xsrange(R)],pointsize=1).show(aspect_ratio=1)
```

```python
grafica1(10^5)
```

```python
def prob_cuerda2(R):
    cont = 0
    cont2 =0
    while cont2 < R:
        x,y = 2*random()-1,2*random()-1
        if x^2+y^2 <= 1:
            cont2 += 1
        if x^2+y^2 < 1/4:
            cont += 1
    return (cont/R).n()
```

```python
prob_cuerda2(10^5)
```

```python
def grafica2(R):
    cont2 =0
    L = []
    while cont2 < R:
        x,y = 2*random()-1,2*random()-1
        if x^2+y^2 <= 1:
            cont2 += 1
            L.append((x,y))
    return points(L,pointsize=1).show(aspect_ratio=1)
```

```python
grafica2(10^5)
```

```python
def prob_cuerda3(R):
    cont = 0
    for muda in xsrange(R):
        P = punto()
        r = random()
        Q = (r*P[0],r*P[1])
        if Q[0]^2+Q[1]^2 < 1/4:
                cont += 1
    return (cont/R).n()
```

```python
prob_cuerda3(10^5)
```

```python
def figura3(R):
    L = []
    for muda in xsrange(R):
        P = punto()
        r = random()
        Q = (r*P[0],r*P[1])
        L.append(Q)
    return points(L,pointsize=1).show(aspect_ratio=1)
```

```python
figura3(10^5)
```

<h3>Cumplea&ntilde;os-Ej-5.3</h3>

```python
def prob_cumple(N,R):
    cont = 0
    for muda in xsrange(R):
        A = set([randint(1,365) for muda in srange(N)])
        if len(A) < N:
            cont += 1
    return (cont/R).n()
```

```python
[prob_cumple(10*k,10^4) for k in srange(1,7)]
```

<h3>Sombreros-Ej-5.2</h3>

```python
def prob_sombreros(N,R):
    cont = 0
    for muda in xsrange(R):
        for k in srange(N):
            x = randint(1,N)
            if x == k:
                cont += 1
                break
    return (cont/R).n()
```

```python
prob_sombreros(20,10^5)
```

```python
prob_sombreros(30,10^5)
```

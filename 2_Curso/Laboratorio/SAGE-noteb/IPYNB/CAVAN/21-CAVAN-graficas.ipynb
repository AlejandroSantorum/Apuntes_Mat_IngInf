<h3>Funciones de una variable</h3>

```python
f(x)=sin(x);f(pi/17);f(pi/17).n()
```

```python
plot(f,(x,-2,2))
```

```python
var('x y');recta = implicit_plot(y-diff(f(x),x)(0)*x,(x,-2,2),(y,-2,2))
```

```python
seno = plot(f(x),(x,-2,2),color=(1,0,0))
```

<h4>Suma de gr&aacute;ficas</h4>

```python
(recta+seno).show(axes=True)
```

<p>Tambi&eacute;n podemos representar curvas planas dadas mediante una relaci&oacute;n entre las variables $x$ e $y$. Decimos que la curva est&aacute; dada en impl&iacute;citas. En este ejemplo la curva pasa dos veces por el origen, y decimos que el origen es un punto singular. <span style="color: #ff0000;">&iquest;Podr&iacute;as encontrar una parametizaci&oacute;n para esta curva?</span></p>

```python
implicit_plot(x^2-y^2+x^3,(x,-5,5),(y,-5,5)).show(axes=True)
```
